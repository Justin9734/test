```python
#import stuff
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as bs
import re
import requests
```

```python
# get request, and put it into viewable variables
URL = "https://docs.google.com/document/u/2/d/e/2PACX-1vQZWtC6NZmslA9Gu2FhQ-qhF9-ECNsrqYYFQYfkASHVN7ztMAEAAAIZeQQr6vlSc5oRyYeRV2A7fEbx/pub?urp=gmail_link"
request = requests.get(URL)
websiteHTML = bs(request.text, 'html5lib')
htmlString = websiteHTML.prettify()
```

```python
#Organize into array

tags = websiteHTML.find_all("tbody")
#print(tags)
array = [[], []]
for x in range(0, len(tags)): # parse through tbody, x is current row
    
    
    for child in tags[x]: #parse through rows
     
        array.append([]) 
        
        for i in child.children: #parse through induvidual things

            placeholder = "" # add everything together
            for string in i.strings:
                placeholder += string
                placeholder += "\n"
            
            array[x].append(placeholder)
           
for x in array:
    print(x)

```

```python
#this cell is for putting the data in a neat lil thing so its visible :D
#cleanup time YAYYYYYY (remove everything that is not part of the table)
placeholderarray = array
i = 0
for x in placeholderarray:
    if len(x) < 3:
        array.pop(i)
    i += 1
newarray = []
ea = 0
#workaround bc the for loops in the previous cell are not for looping
for x in range(0, len(array)): #go through array

    for i in range(0, len(array[x])): # go through the 2nd dimension

        if ea == 0: # more workarounds, grab a group of 3 and put it in the new list, the 3 being date, day of week, and anouncements

            newarray.append([array[x][i+1], array[x][i], array[x][i+2]]) #a bit silly here, swaped for formatting purposes
        ea += 1
        ea = ea % 3
array = newarray
# replace no anouncements with n/a bc it looks nicer
for x in range(0, len(array)):
    if array[x][2]== "":
        array[x][2] = "N/A"
for x in array:
    print(x)

```
